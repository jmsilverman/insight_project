# Population Info
### From US Census (2010)
### Puts together a bunch of data from 3 different census tables:
* census_tract
* population: total population, both sexes
* population_percent_male: percentage of total population that is male
* population_percent_female: percentage of total population that is female
* age_med: median age for everyone
* age_med_male: median age for males
* age_med_female: median age for females
* minor_percent: percent of population under 18 years
* adult_percent: percent of population over 18 years

### Makes some histograms of the above data
### All of this gets output to data_final/sd_population_cleaned.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in age info
ages = pd.read_csv('data/American_FactFinder_census/DEC_10_SF1_P13_with_ann.csv',skiprows=[1],dtype={'GEO.id2': str})
# count how many ages we got
print ages['GEO.id2'].count()
ages
```

```python
# plot age info
plots = ages.plot.hist(bins=20,figsize=(13, 10),subplots=True)
plt.xlabel('Median Age')
plots[0].legend(('All',))
plots[1].legend(('Male',))
plots[2].legend(('Female',))
plt.show()
```

```python
# read in minor vs. adult info
minors = pd.read_csv('data/American_FactFinder_census/DEC_10_SF1_P16_with_ann.csv',skiprows=[1],dtype={'GEO.id2': str})
# count how many rows we got
print minors['GEO.id2'].count()
minors
```

```python
# plot minor vs. adult info
plt.figure(figsize=(13, 10))
ax1 = plt.subplot(211)
young = minors.D002[:-1]/minors.D001[:-1]
plt.hist(young.dropna(), 20)
plt.xlabel('Percent Minors')
plt.ylabel('# of Census Blocks')
ax2 = plt.subplot(212, sharex=ax1)
old = minors.D003[:-1]/minors.D001[:-1]
plt.hist(old.dropna(), 20)
plt.xlabel('Percent Adults')
plt.ylabel('# of Census Blocks')
plt.show()
```

```python
# read in male/female info
mf = pd.read_csv('data/American_FactFinder_census/DEC_10_SF1_QTP1_with_ann.csv',skiprows=[1],dtype={'GEO.id2': str})
# count how many rows we got
print mf['GEO.id2'].count()
mf
```

```python
# plot male/female ratio
plt.figure(figsize=(15, 10))
ax1 = plt.subplot(211)
male = mf['SUBHD0102_S01']/mf['SUBHD0101_S01']
plt.hist(male.dropna(), 30)
plt.xlabel('Percent Male')
plt.ylabel('# of Census Blocks')
ax2 = plt.subplot(212, sharex=ax1)
fem = mf['SUBHD0103_S01']/mf['SUBHD0101_S01']
plt.hist(fem.dropna(), 30)
plt.xlabel('Percent Female')
plt.ylabel('# of Census Blocks')
plt.show()
```

```python
# make sure all 3 tables have the same rows in the same order by checking their census tract numbers
print sum(mf['GEO.id2']!=minors['GEO.id2'])
print sum(mf['GEO.id2']!=ages['GEO.id2'])
```

```python
# combine all info we want to keep and give them reasonable key names
result = pd.concat([ages['GEO.id2'].str[5:], \
                    mf['SUBHD0101_S01'], mf['SUBHD0102_S01']/mf['SUBHD0101_S01'], mf['SUBHD0103_S01']/mf['SUBHD0101_S01'], \
                    ages['D002'], ages['D003'], ages['D004'], \
                    minors['D002']/minors['D001'], minors['D003']/minors['D001']], axis=1, \
                    keys=['census_tract','population','population_percent_male','population_percent_female', \
                         'age_med','age_med_male','age_med_female','minor_percent','adult_percent'])
#result.to_csv('data_final/sd_population_cleaned.csv',na_rep='0',index=False)
result
```
